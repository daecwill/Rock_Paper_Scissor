# ROCK PAPER SCISSORS
Create a program where the user can play a game of rock, paper, scissors against a computer. Set a play choice (rock, paper, or scissors) for the computer player and during the program, request for the user to input their play choice.

The rules in Rock, Paper, Scissors are:

Rock beats scissors
Scissors beats paper
Paper beats rock
If the user wins, then display "You win!" If they lose against the computer, then display "You lose." If the computer and the user have the same play choice, display "It's a draw."

BONUS: Randomize the play choice for the computer (hint: Python has a library called random). Also ask the user if they would like to play again. If they do, start the game again but if not, thank the user for playing the game.
CHALLENGE = KEEP SCORE


```python
print("Welcome to the Rock Paper Scissors Game!\n")

import random as rand 

score = []
pc_score = 0
player_score = 0
pcs_print = ""
ps_print = ""

keep_playing = "y"
while keep_playing == "y" :
    player_select = str(input("Select your item (r for rock, p for paper, or s for scissors) :  "))
    ps = player_select.lower().strip()

    match = ""
 
    if ps != "r" and ps != "p" and ps != "s" :
        print("Please select the proper item and try again")
        
    #pretty up what the player picked , but keep simple for code
    if ps == "r":
        ps_print = "rock"
    elif ps == "p" :
        ps_print = "paper"
    else:
        ps_print = "scissors"   

    # pc chooses randomly
    pc_select = rand.randint(1, 3) 
    if pc_select == 1: 
        pcs = "r"
    elif pc_select == 2:
        pcs = "p"
    elif pc_select == 3:
        pcs = "s" 
    
    #pretty up what the computer picked, but keep simple for code
    if pcs == "r":
        pcs_print = "rock"
    elif pcs == "p" :
        pcs_print = "paper"
    else:
        pcs_print = "scissors"
    
    print()
    print(f"You picked: {ps_print}")
    print(f"Computer picked: {pcs_print}")

    # lets compare
    # ps = what the player selected
    # pcs = what the PC randomly selected
    if ps == "r":
        if pcs == "r":
            match = "it's a draw."
        elif pcs == "p":
            match = "You Lose."
        elif pcs == "s" :
            match = "You Win!"
    elif ps == "p" :
        if pcs == "r":
            match = "You Win!"
        elif pcs == "p":
            match = "It's a draw."
        elif pcs == "s" :
            match = "You Lose."    
    elif ps == "s" :
        if pcs == "r":
            match = "You Lose."
        elif pcs == "p":
            match = "You Win!"
        elif pcs == "s" :
            match = "it's a draw."
    else:
        match = "Inform your IT department"


    print()
    print(match)
    print()
    
    # keep score
    if match == "You Lose.":
        pc_score += 1
    elif match == "You Win!":
        player_score += 1
        
    
    kp = input("Would you like to play again? (y / n)  ") 
    print()
    if kp != "y":
        
        print()
        print(f"Before you go ... \nYour score is : {player_score} \nComputer Score is : {pc_score}")
        print()
        print("Thanks for playing.")
        
        keep_playing = "n"
        break
        
```
